# Знакомство с jupyter

Welcome в так называемый блокнот системы jupyter =) Такими блокнотами мы будем пользоваться для увеличения интерактивности курса и изучения возможностей системы jupyter!  

Во-первых, не пугаемся) Блокнот состоит из блоков, каждый из которых может представлять собой разметку (Markdown) или код. Код можно писать, выполнять и получать мгновенный результат, если код предполагает вывод.

На верху у вас есть панелька, с помощью которой вы можете управлять выполняемым кодом  
![Панель]("img/Panel.png")  

Слева от вас есть курсор (синенький), он сообщает на какой ячейке блокнота вы находитесь. Нажатие на кнопку `Play` выполняет ячейку кода (если ячейка содержит разметку - он просто приводит ее в результирующий вид).  

> Важно отметить порядок выполнения ячеек: в том порядке, как вы выполняете ячейки кода, в таком порядке будут записываться данные!  
То есть, порядком выполнения ячеек кода вы выстраиваете порядок результирующего кода. Надеюсь не стоит говорить о том, что выполнять ячейку чтения переменной, если никакая ячейка ранее не задала эту переменную - бессмысленно? =)

Ядро (kernel), в котором выполняется код, является как бы вашей сессией. Если вы перезапускаете ядро - все данные сбрасываются и вам требуется выполнить коды внутри блоков заново, чтобы все данные обработались заново.


Теперь попробуем немного поработать с кодом. Для начала простой вывод на языке `Python3`. Переведите курсор на ячейку с кодом и нажмите кнопку `Play` (`Run the selected cells and advance`)

```python
print('Hello')
```

Как видно, под блоком кода появился вывод. Наверное такой мы и ожидали =)  

> Этот вывод можно очистить, правой кнопкой мыши на ячейку и жмите `Clear outputs`. Аналогично можно очистить весь вывод блокнота `Clear all outputs`.  

Теперь немного сложнее, математика, попробуем задать переменную

```python
exVar = 4 + 2
```

```python
exVar
```

Опа, то есть, при написании имени переменной, можно получить ее вывод, интересно.. А что если сделать все в одном блоке?
